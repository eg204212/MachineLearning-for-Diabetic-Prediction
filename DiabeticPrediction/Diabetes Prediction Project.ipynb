# Diabetes Prediction Project

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.shape # 768 rows and 9 columns
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Outcome'].value_counts()
```

# 0 - non diabetic
# 1 - diabetic

```python
# drop the outcome column, 

x = df.drop(columns = 'Outcome' , axis =1)
y = df['Outcome']
```

```python
x # without outcome
```

```python
y # outcome
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
standardized_data
```

```python
x = standardized_data
```

```python
x # final x
```

```python
y # final y
```

# training and testing part

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train.shape # 80% 
```

```python
x_test.shape # 20% for testing(0.2)
```

# Check the accuracy of train and test model

```python
# classification problem to find the person who has diabetes or not
clf = svm.SVC(kernel='linear')
```

```python
clf.fit(x_train,y_train)
```

```python
# accuracy for the training data

x_train_prediction = clf.predict(x_train)
accuracy_score(x_train_prediction,y_train)
```

```python
# accuracy of the test data

x_test_prediction = clf.predict(x_test)
accuracy_score(x_test_prediction,y_test)
```

```python
# based on the features predict person has or not
# input data for the 8 columns for a random person


input_sample = (0.5,100,55,14,33.5,5,9,56)
```

```python
input_np_array = np.asarray(input_sample)
```

```python
input_np_array_reshaped = input_np_array.reshape(1,-1)
```

```python
# standardize the input data

std_data = scaler.transform(input_np_array_reshaped)
```

```python
std_data
```

```python
prediction = clf.predict(std_data)
prediction  # prediction is 1 , it means person is diabetes
```

```python
if (prediction[0]==0):
    print("Person does not suffer from diabetic")
else:
    print("Person suffers from diabetic")
```

```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np  # You need to import NumPy for array manipulation
from IPython.display import clear_output

# Assuming you have a trained model named `model` for predicting diabetes
# Make sure you have defined the predict_diabetes function using this model

# Define a function to predict diabetes based on input features
def predict_diabetes(features):
    # Your prediction logic here using your trained model
    # For demonstration, let's assume it's a simple threshold-based prediction
    threshold = 0.5  # You may adjust the threshold as per your model's requirements
    prediction = np.where(features > threshold, 1, 0)
    return prediction

# Create input widgets for sample data
pregnancies_input = widgets.IntSlider(description='Pregnancies:', min=0, max=17, value=0)
glucose_input = widgets.IntSlider(description='Glucose:', min=0, max=200, value=100)
blood_pressure_input = widgets.IntSlider(description='Blood Pressure:', min=0, max=122, value=72)
skin_thickness_input = widgets.IntSlider(description='Skin Thickness:', min=0, max=99, value=23)
insulin_input = widgets.IntSlider(description='Insulin:', min=0, max=846, value=79)
bmi_input = widgets.FloatSlider(description='BMI:', min=0, max=67, value=32.0)
dpf_input = widgets.FloatSlider(description='Diabetes Pedigree Function:', min=0.0, max=2.42, value=0.3725)
age_input = widgets.IntSlider(description='Age:', min=21, max=81, value=29)

# Define a function to handle prediction when the button is clicked
def on_button_clicked(b):
    clear_output(wait=True)  # Clear the output area, wait=True ensures that the output is cleared immediately
    features = [
        pregnancies_input.value,
        glucose_input.value,
        blood_pressure_input.value,
        skin_thickness_input.value,
        insulin_input.value,
        bmi_input.value,
        dpf_input.value,
        age_input.value
    ]
    prediction = predict_diabetes(np.array(features))  # Use your predict_diabetes function
    if prediction[0] == 0:
        print("The person does not have diabetes.")
    else:
        print("The person has diabetes.")
        
# Create a button widget
button = widgets.Button(description="Predict")

# Attach the button click event to the function
button.on_click(on_button_clicked)

# Display the widgets
display(
    pregnancies_input,
    glucose_input,
    blood_pressure_input,
    skin_thickness_input,
    insulin_input,
    bmi_input,
    dpf_input,
    age_input,
    button
)
```
